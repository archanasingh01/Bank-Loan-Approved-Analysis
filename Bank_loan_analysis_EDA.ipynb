```python
# home ownership status
person_home_ownership_count = df["person_home_ownership"].value_counts()
plt.bar(person_home_ownership_count.index, person_home_ownership_count.values)
plt.title("Home ownership status")
plt.xlabel("Home ownership")
plt.ylabel("number of applications")
plt.show()
```

```python
Interpretation - Applicants Living in rented homes submitted the highest number of loan applications, while applicants who own thier home homes
                 submitted comparitively fewer loan applications.

Buisness Insight - The majority of loan demand in this dataset comes from people living in rented homes. This suggests that renters are the primary 
                   customer segment applying for loans.
```

```python
# default on file
default_on_file_count = df["cb_person_default_on_file"].value_counts()
plt.bar(default_on_file_count.index, default_on_file_count.values)
plt.title("Default on file")
plt.xlabel("Default on file")
plt.ylabel("number of applications")
plt.show()
```

```python
# Bivariate Analysis
```

```python
# Grouped Bar Chart: Loan status vs Loan intent
loan_intent_status= df.groupby(["loan_intent", "loan_status"]).size().unstack()
loan_intent_status.plot(kind="bar", figsize=(10,6))
plt.title("loan status by Loan intent")
plt.xlabel("Loan intent")
plt.ylabel("number of applications")
plt.legend(["Rejected (0)", "Approved (1)"])
plt.xticks(rotation=45)
plt.show()
```

```python
Interpretation - Education loans generate the highest demand among applicants. However, despite the large number of applications, rejected applications
                 still outnumber approved applications. A similiar trend is observed across other loan purposes, suggesting that criteria regardless
                 of loan approval.

Buisness insights - Education loans represents the largest demand among applicants.
                    Regardless of loan purpose, rejection counts are consistently higher than approval counts, indicating that loan approval is 
                    selective and likely depends on additional applicant factors such as income, credit history, or loan grade.
```

```python
#  Loan Status vs Home ownership
home_ownership_status = df.groupby(["person_home_ownership", "loan_status"]).size().unstack()
home_ownership_status.plot(kind="bar")
plt.title("Loan status vs Home ownership")
plt.xlabel("Home ownership")
plt.ylabel("Number of applications")
plt.legend(["Rejected (0)", "Approved (1)"])
plt.show()
```

```python
Interpretation -  Applicants living in rented homes and mortgaged homes have the highest number of loan applications. Applicants in the "own" and
                  "other" categories have comparitively fewer loan applications. In all categories, rejected applications are higher than approved 
                  applications.

Buisness Insight - Most loan demand comes from applicants living in rented or mortgaged homes. However, across all home ownership categories, rejected 
                   applications exceed approved applications, suggesting that home ownership alone is not sufficient for loan approval.
```

```python
# Loan status vs Loan grade
loan_Grade_status= df.groupby(["loan_grade", "loan_status"]).size().unstack()
loan_Grade_status.plot(kind="bar", figsize=(10,6))
plt.title("loan status by Loan grade")
plt.xlabel("Loan grade")
plt.ylabel("number of applications")
plt.legend(["Rejected (0)", "Approved (1)"])
plt.xticks(rotation=0)
plt.show()
```

```python
Interpretation - Loan Grades A and B have the highest number of loan applications.
                 Loan Grades F and G have the lowest number of loan applications.
                 For Grades A, B, and C, rejected applications are higher than approved applications.
                 For Grades D, E, F, and G, approved applications are higher than rejected application.

Buisness Insight - Loan applications are concentrated in Grades A and B, indicating that most applicants belong to these credit grades. The relationship
                   between loan grade and loan approval varies across grades, suggesting that loan grade alone does not determine the bank's lending 
                   decision. Other factors such as income, previous default history, employment length, and loan amount may also influence approval.
```

```python
# Correlation Analysis
```

```python
numerical_df = df.select_dtypes(include="number")
corr_matrix = numerical_df.corr()
corr_matrix
```

```python
Interpretation

Age and Credit History Length have a strong positive correlation (0.86), meaning older applicants generally have a longer credit history
    .
Loan Amount and Loan Percent Income have a moderate positive correlation (0.57), indicating that larger loan amounts usually make up a larger 
percentage of the applicant's income.

Loan Status and Loan Interest Rate show a moderate positive correlation (0.32), suggesting that loan approval status has some relationship with the 
interest rate.
    
Loan Status and Loan Percent Income also have a moderate positive correlation (0.38), indicating that the loan-to-income ratio is related to loan 
approval decisions.
    
Person Income and Loan Amount have a weak positive correlation (0.27), meaning applicants with higher income tend to take slightly larger loans.
    
Most of the remaining variables have very weak correlations (close to 0), indicating little or no linear relationship.
```

```python
4. Key Business Insights

Most loan applicants belong to the 20–35 years age group.

Income distribution is positively skewed, with a few high-income applicants acting as outliers.
    
Most applicants request small to medium-sized loan amounts.
    
The majority of loans have interest rates between 10% and 13%.
    
Most applicants have relatively short employment histories.
    
Education is the most common purpose for applying for a loan.
    
Applicants living in rented homes represent the largest group of loan applicants.
    
Most applicants have no previous loan default, indicating a generally good credit history.
    
Loan Grades A and B contain the highest number of applicants.
    
Across most categories, rejected loan applications exceed approved applications, suggesting that loan approval depends on multiple applicant 
characteristics.
    
Correlation analysis shows relationships between important numerical variables, helping identify factors associated with loan amount and applicant 
income.
    
Applicant age is strongly associated with credit history, which may help lenders assess applicant reliability.
    
Loan amount and loan-to-income ratio are important financial indicators and should be considered together during loan evaluation.
    
Income alone does not strongly determine loan amount, suggesting that lenders also consider other applicant characteristics.
    
Since most variables have weak correlations, loan approval depends on multiple factors rather than a single variable.
```
